```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../ncbi_590_data.csv')
df = df[df['is_usa_based'] == 1]
```

```python
#Vegetable
#Fruit
#Chicken
#Beef
#Pork
#Turkey
```

```python
#add = pd.read_csv('ncbi_to_nors_source_match.csv')
add = pd.read_csv('ncbi_salmonella_fruit_veg_pol.csv')
df['primary_mode_fruit_veg_pol'] = df['isolation_source'].copy()
```

```python
value_counts_index = df['primary_mode_fruit_veg_pol'].value_counts().index
for i, top_value in enumerate(value_counts_index):
    print(f"Top value {i}: {top_value}")
    user_input = input(f"Enter a number for '{top_value}' (1-6): ")
    if user_input == "1":
        user_input = "Vegetable"


    elif user_input == "2":
        user_input = "Fruit"


    elif user_input == "3":
        user_input = "Chicken"


    elif user_input == "4":
        user_input = "Beef"


    elif user_input == "5":
        user_input = "Pork"

        
    elif user_input == "6":
        user_input = "Turkey"
    elif user_input == "0":
        break
    else:
        user_input = "Other/Unknown"
    add = pd.concat([add, pd.DataFrame({'before': [top_value], 'after': [user_input]})], ignore_index=True)
    add.to_csv('ncbi_salmonella_fruit_veg_pol.csv', index=False)
```
