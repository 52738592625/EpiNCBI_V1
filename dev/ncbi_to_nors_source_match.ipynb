```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../ncbi_562_data.csv')
df_other = pd.read_csv('../NORS_20250702.csv')
df_other['Primary Mode'].value_counts()
```

```python
add = pd.read_csv('ncbi_to_nors_source_match.csv')
df['nors_primary_mode_map'] = df['isolation_source'].copy()
```

```python
#replace the entries in df['nors_primary_mode_map'] with the cooresponding values in ncbi_to_nors_source_match.csv
df_map = pd.read_csv('ncbi_to_nors_source_match.csv')
#rename the columns to isolation_source and Primary Mode
df_map.columns = ['isolation_source', 'Primary Mode']
df['nors_primary_mode_map'] = df['nors_primary_mode_map'].replace(df_map.set_index('isolation_source')['Primary Mode'].to_dict())
```

```python

#write a for loop that prints the top value of df['nors_primary_mode_map'].value_counts() asks the user to input a number from 1 to 5,
#and then replaces the top value with the user input
value_counts_index = df['nors_primary_mode_map'].value_counts().index
for i, top_value in enumerate(value_counts_index):
    print(f"Top value {i}: {top_value}")
    user_input = input(f"Enter a number for '{top_value}' (1-6): ")
    if user_input == "1":
        user_input = "Person-to-person"


    elif user_input == "2":
        user_input = "Food"


    elif user_input == "3":
        user_input = "Indeterminate/unknown"


    elif user_input == "4":
        user_input = "Water"


    elif user_input == "5":
        user_input = "Animal contact"

        
    elif user_input == "6":
        user_input = "Environmental contamination other than food/water"
    else:
        break
    add = pd.concat([add, pd.DataFrame({'before': [top_value], 'after': [user_input]})], ignore_index=True)


```
