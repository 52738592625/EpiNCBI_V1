```python
import pandas as pd
df = pd.read_csv('../ncbi_590_data.csv')
#display max rows
pd.set_option('display.max_rows', None)

df.head()
```

```python
df['collection_date'].value_counts()
```

```python
#If the 5th character is a -, then try to extract the 6th and 7th characters as the month
df['month'] = df['collection_date'].apply(lambda x: x[5:7] if len(x) > 5 and x[4] == '-' else None)
#convert month to numeric
df['month'] = pd.to_numeric(df['month'], errors='coerce')

```
