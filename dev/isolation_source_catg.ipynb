```python
import pandas as pd
import os
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
ecoli = pd.read_csv('../ncbi_562_data.csv')
salmonella = pd.read_csv('../ncbi_590_data.csv')
```

```python
ecoli = ecoli[ecoli['is_usa_based'] == 1]
ecoli['collection_year'] = pd.to_numeric(ecoli['collection_year'], errors='coerce')
ecoli = ecoli[ecoli['collection_year'] >= 2009]
ecoli = ecoli[ecoli['month'].notna()]
ecoli['collection_year'] = ecoli['collection_year'].astype(str)
ecoli['collection_year'] = ecoli['collection_year'].str[:4]
ecoli['month'] = ecoli['month'].astype(int)
ecoli['month'] = ecoli['month'].apply(lambda x: f"{x:02d}")
ecoli['year_and_month'] = ecoli['collection_year'] + '-' + ecoli['month']

salmonella = salmonella[salmonella['is_usa_based'] == 1]
salmonella['collection_year'] = pd.to_numeric(salmonella['collection_year'], errors='coerce')
salmonella = salmonella[salmonella['collection_year'] >= 2009]
salmonella = salmonella[salmonella['month'].notna()]
salmonella['collection_year'] = salmonella['collection_year'].astype(str)
salmonella['collection_year'] = salmonella['collection_year'].str[:4]
salmonella['month'] = salmonella['month'].astype(int)
salmonella['month'] = salmonella['month'].apply(lambda x: f"{x:02d}")
salmonella['year_and_month'] = salmonella['collection_year'] + '-' + salmonella['month']
```

```python
ecoli['source_category'] = ecoli['isolation_source'].copy()
salmonella['source_category'] = salmonella['isolation_source'].copy()
```

```python
if os.path.exists('../lookup_tables/ecoli_isolation_source_catg_map.csv'):
    ecoli_map = pd.read_csv('../lookup_tables/ecoli_isolation_source_catg_map.csv')
    ecoli['source_category'] = ecoli['source_category'].replace(ecoli_map.set_index('isolation_source')['source_category'].to_dict())
else:
    ecoli_map = pd.DataFrame(columns=['isolation_source', 'source_category'])
if os.path.exists('../lookup_tables/salmonella_isolation_source_catg_map.csv'):
    salmonella_map = pd.read_csv('../lookup_tables/salmonella_isolation_source_catg_map.csv')
    salmonella['source_category'] = salmonella['source_category'].replace(salmonella_map.set_index('isolation_source')['source_category'].to_dict())
else:
    salmonella_map = pd.DataFrame(columns=['isolation_source', 'source_category'])
```

```python
value_counts_index = ecoli['source_category'].value_counts().index
for i, top_value in enumerate(value_counts_index):
    user_input = input(f"'{top_value}")
    if user_input == "1":
        user_input = "Vegetable"


    elif user_input == "2":
        user_input = "Fruit"


    elif user_input == "3":
        user_input = "Chicken"


    elif user_input == "4":
        user_input = "Beef"


    elif user_input == "5":
        user_input = "Pork"

        
    elif user_input == "6":
        user_input = "Turkey"

    elif user_input == "7":
        user_input = "Human"

    elif user_input == "8":
        user_input = "Water"

    elif user_input == "9":
        user_input = "Farm"
    
    elif user_input == "0":
        user_input = "Community"

    elif user_input == "`":
        user_input = "Other/Unknown"
    
    elif user_input == "-":
        break
    ecoli_map = pd.concat([ecoli_map, pd.DataFrame({'isolation_source': [top_value], 'source_category': [user_input]})], ignore_index=True)
    ecoli_map.to_csv('../lookup_tables/ecoli_isolation_source_catg_map.csv', index=False)
```
