```python
import pandas as pd
df = pd.read_csv('NORS_20241209.csv')
df.head(10)
```

```python
years = range(1971, 2025)
months = range(1, 13)
#Create a new dataframe with all the year, month pairs
df1 = pd.DataFrame([(y, m) for y in years for m in months], columns=['Year', 'Month'])
#Number the unqiue month, year pairs and store in a new column called seq_num
df1['seq_num'] = df1.groupby(['Year', 'Month']).ngroup()
df1.head(10)
```

```python
#sort data by year and month in ascending order
df2 = df.sort_values(by=['Year', 'Month'], ascending=[True, True]).reset_index(drop=True)
#create a seq_num column based off seq_num in df1
df2['seq_num'] = 0
for i in range(len(df1)):
    df2.loc[(df2['Year'] == df1['Year'][i]) & (df2['Month'] == df1['Month'][i]), 'seq_num'] = df1['seq_num'][i]
    print(i)
df2.head(10)
```
