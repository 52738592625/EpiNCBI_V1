```python
#Filters out the non-US states, and edits geo_loc_name into 'state' column to have only US state names,
!pip install numpy
!pip install matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('10__data_unstacked.csv', dtype=str)
```

```python
keep_list_us = []
for i in df['geo_loc_name'].unique().tolist()[1:]:
    if 'USA' in i:
        keep_list_us.append(i)
```

```python
#keep the rows with keep_list_us in df['geo_loc_name]
df_us = df[df['geo_loc_name'].isin(keep_list_us)]
```

```python
df_us['state'] = df_us['geo_loc_name'].copy()
list_of_states_ = ['USA: AL', 'USA: AK', 'USA: AZ', 'USA: AR', 'USA: CA', 'USA: CO', 'USA: CT', 'USA: DE', 'USA: FL', 'USA: GA', 'USA: HI', 'USA: ID', 'USA: IL', 'USA: IN', 'USA: IA', 'USA: KS', 'USA: KY', 'USA: LA', 'USA: ME', 'USA: MD', 'USA: MA', 'USA: MI', 'USA: MN', 'USA: MS', 'USA: MO', 'USA: MT', 'USA: NE', 'USA: NV', 'USA: NH', 'USA: NJ', 'USA: NM', 'USA: NY', 'USA: NC', 'USA: ND', 'USA: OH', 'USA: OK', 'USA: OR', 'USA: PA', 'USA: RI', 'USA: SC', 'USA: SD', 'USA: TN', 'USA: TX', 'USA: UT', 'USA: VT', 'USA: VA', 'USA: WA', 'USA: WV', 'USA: WI', 'USA: WY']
list_of_states = ['USA:AL','USA:AK','USA:AZ','USA:AR','USA:CA','USA:CO','USA:CT','USA:DE','USA:FL','USA:GA','USA:HI','USA:ID','USA:IL','USA:IN','USA:IA','USA:KS','USA:KY','USA:LA','USA:ME','USA:MD','USA:MA','USA:MI','USA:MN','USA:MS','USA:MO','USA:MT','USA:NE','USA:NV','USA:NH','USA:NJ','USA:NM','USA:NY','USA:NC','USA:ND','USA:OH','USA:OK','USA:OR','USA:PA','USA:RI','USA:SC','USA:SD','USA:TN','USA:TX','USA:UT','USA:VT','USA:VA','USA:WA','USA:WV','USA:WI','USA:WY']
states = ['Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Delaware','Florida','Georgia','Hawaii','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana','Maine','Maryland','Massachusetts','Michigan','Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico','New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont','Virginia','Washington','West Virginia','Wisconsin','Wyoming']
df_us['state'] = df_us['state'].replace(['USA: Texas','USA:Texas', 'USA: TX'], 'USA:TX')
df_us['state'] = df_us['state'].replace(['USA: California'], 'USA:CA')
df_us['state'] = df_us['state'].replace(['USA: NC'], 'USA:NC')
df_us['state'] = df_us['state'].replace(['USA: San Diego, CA'], 'USA:CA')
df_us['state'] = df_us['state'].replace(['USA: MO'], 'USA:MO')
df_us['state'] = df_us['state'].replace(['USA: Maryland', 'USA: MD'], 'USA:MD')
df_us['state'] = df_us['state'].replace(['USA: NM'], 'USA:NM')
df_us['state'] = df_us['state'].replace(['USA: New York State', 'USA: New York state', 'USA: NY', 'USA: New York'], 'USA:NY')
df_us['state'] = df_us['state'].replace(['USA: Washington', 'USA: Washington state'], 'USA:WA')
df_us['state'] = df_us['state'].replace(['USA: Nebraska', 'USA:Nebraska'], 'USA:NE')
df_us['state'] = df_us['state'].replace(['USA:Indiana'], 'USA:IN')
df_us['state'] = df_us['state'].replace(['USA: GA', 'USA: Georgia'], 'USA:GA')
df_us['state'] = df_us['state'].replace(['USA: PA'], 'USA:PA')
df_us['state'] = df_us['state'].replace(['USA: Minnesota'], 'USA:MN')
for i in range(len(list_of_states)):
    df_us['state'] = df_us['state'].replace([list_of_states_[i]], list_of_states[i])



df_us['state'].value_counts()
```

```python
for i in range(len(list_of_states)):
    df_us['state'] = df_us['state'].replace([list_of_states[i]], states[i])
df_us['state'].value_counts()
```

```python
n = 0
for i in range(len(df_us)):
    if df_us['state'].iloc[i] not in states:
        df_us['state'].iloc[i] = 'Uncategorized'
        if n % 10000 == 0:
            print(n)
        n += 1
```
