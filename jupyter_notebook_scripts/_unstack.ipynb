```python
df = pd.read_csv('ecoli_data.tsv', sep='\t')
```

```python
#rename columns
df.columns = ['genome', 'attribute', 'attribute_value']
```

```python
#pivot table, handle duplicates with first
df = df.pivot_table(index='genome', columns='attribute', values='attribute_value', aggfunc='first').reset_index()
df.columns.name = None
```

```python
df.to_csv('10__data_unstacked.csv', index=False)
```
