```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('data.tsv', sep='\t')
```

```python
df = df.iloc[:, 0:3]
df.columns = ['sample', 'attribute', 'value']
df.head(10)
```

```python
#groupby sample
df_new = df.groupby('sample')
df_new = df_new['attribute'].value_counts().unstack()
df_new.columns.name = 'sample'
df_new = df_new.reset_index()
df_new.columns.name = None
unique_samples = df_new['sample'].unique()


```

```python
for sample in unique_samples:
    df_temp = df[df['sample'] == sample]
    unique_attributes = df_temp['attribute'].unique()
    for attribute in unique_attributes:
        value = df_temp[df_temp['attribute'] == attribute]['value'].values[0]
        df_new.loc[df_new['sample'] == sample, attribute] = value
df_new.head(10)
```
