```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
n=0
df = pd.read_csv('16_states_dates_sero.csv', dtype = str)
df['strain'] = df['strain'].fillna('missing')
df['strain_category'] = df['strain']
for i in range(len(df)):
    if 'ESBL' in df.loc[i, 'strain']:
        df.loc[i, 'strain_category'] = '1'
    else:
        if 'missing' not in df.loc[i, 'strain']:
            df.loc[i, 'strain_category'] = '0'
    if n % 10000 == 0:
        print(n)
    n+=1

```

```python
df_final_ish = df[['state', 'collection_year', 'strain_category', 'isolation_source', 'new_isolation_source', 'serotype', 'serotype_categorized', 'strain']]
```

```python
df_final_ish.to_csv('17_columns_we_want.csv', index = False)
del [df_final_ish, df]
```
