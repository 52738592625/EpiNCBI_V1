```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('15_states_dates.csv', dtype = str)
```

```python
df['serotype'].fillna('missing', inplace=True)
```

```python
n = 0
df['serotype_categorized'] = df['serotype']
df['serotype_categorized'] = df['serotype_categorized'].replace(['Missing','not collected','blank'], 'missing')
df['serotype_categorized'] = df['serotype_categorized'].replace('O157: H7', 'O157:H7')
df['serotype_categorized'] = df['serotype_categorized'].replace('O157:NM', 'O157')
df['serotype_categorized'] = df['serotype_categorized'].replace('O157:H48', 'O157')
df['serotype_categorized'] = df['serotype_categorized'].replace('O157: nonmotile', 'O157')
df['serotype_categorized'] = df['serotype_categorized'].replace('Shiga Toxin Positive, non O157:H7', 'other')
for i in range(len(df)):
    n += 1
    x = df.loc[i, 'serotype_categorized']
    if x != 'missing' and 'O157' not in x:
        df.loc[i, 'serotype_categorized'] = 'other'
    if n % 5000 == 0:
        print(n)
df['serotype_categorized'].value_counts()


```

```python
df.to_csv('16_states_dates_sero.csv', index=False)
del df
```
