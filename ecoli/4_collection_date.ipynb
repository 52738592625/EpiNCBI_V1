```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('14_isolation_source.csv', dtype=str)
```

```python
#df['collection_date'].value_counts()
```

```python
n = 0
df['collection_date'] = df['collection_date'].fillna('missing')
df['collection_year'] = df['collection_date']
years=[str(x) for x in list(range(1900,2025))]
df['collection_year'] = df['collection_year'].replace('Missing', 'missing').replace('Unknown', 'missing').replace('unknown', 'missing').replace('none','missing').replace('not applicable','missing').replace('not collected','missing').replace('not provided','missing').replace('Not collected','missing').replace('MISSING','missing').replace('not available', 'missing')
#if the right four characters are in years, then keep them, otherwise replace with missing
for i in range(len(df)):
    n += 1
    r = df.loc[i,'collection_date'][-4:]
    l = df.loc[i,'collection_date'][0:4]
    if r in years:
       df.loc[i,'collection_year'] = r
    elif l in years:
        df.loc[i,'collection_year'] = l
    if n % 5000 == 0:
        print(n)


df['collection_year'].value_counts()
```

```python
df.to_csv('15_states_dates.csv', index=False)
```
